```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 1, 1001)
```

```python
a = 3
b = 4
```

```python
Y = a * X + b
```

```python
plt.plot(X, Y, ".");
```

```python
E = np.random.randn(1001)
```

```python
Y1 = Y + E
```

```python
plt.plot(X, Y1, ".")
```

```python
A = np.vstack([X, np.ones_like(X)]).T
A
```

```python
ATA = A.T @ A
ATA
```

```python
ATy = A.T @ Y1
ATy
```

```python
theta = np.linalg.solve(ATA, ATy)
theta
```

```python
astar, bstar = theta
```

```python
plt.plot(X, Y1, ".")
plt.plot(X, astar*X + bstar);
```

```python
X = np.linspace(0, 1, 2001)
Y = X + 4 + np.random.randn(2001)
```

```python
X.shape
```

```python
def f(a, b, X, Y):
    Y_star = a*X + b
    E = Y - Y_star
    return np.dot(E, E)
```

```python
f(3, 3, X, Y)
```

```python
A = np.linspace(-2, 4, 201)
B = np.linspace(1, 7, 201)
```

```python
AA, BB = np.meshgrid(A, B)
```

```python
F = np.empty_like(AA)
for i in range(AA.shape[0]):
    for j in range(AA.shape[1]):
        a = AA[i, j]
        b = BB[i, j]
        d = f(a, b, X, Y)
        F[i, j] = d
```

```python
plt.contour(AA, BB, F, 20);
```
