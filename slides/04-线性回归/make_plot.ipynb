```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.top'] = False

mpl.rcParams['font.family'] = 'sans-serif'
mpl.rcParams["font.sans-serif"]=["Hiragino Sans GB"] #设置字体
mpl.rcParams["axes.unicode_minus"]=False #该语句解决图像中的“-”负号的乱码问题
```

```python
X = [1000, 1500]
Y = [20, 35]
a = 0.03
b = -10

plt.figure(facecolor="white")
I = np.linspace(800, 1700)
J = I*a + b
plt.plot(I, J, "--", label="模型")

plt.scatter(X, Y, marker="X", c="C1", label="数据", zorder=2)

#for p in zip(X, Y):
#plt.annotate(str(p), p)

plt.title("线性回归（两点间）")
plt.xlabel("人口")
plt.ylabel("出行量")
plt.legend()
plt.savefig("aaa.png", dpi=300)
```
