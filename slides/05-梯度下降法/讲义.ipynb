```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 2, 3])
Y = np.array([4, 7, 6])
```

```python
plt.plot(X, Y, ".")
plt.gca().set_aspect('equal')
```

```python
a = np.linspace(0, 4, 101)
b = np.linspace(0, 4, 101)
```

```python
A, B = np.meshgrid(a, b)
```

```python
def loss(x, y, A, B):
    y_hat = A*x+B
    e = y_hat - y
    return e**2
```

```python
l = loss(X[0], Y[0], A, B)
#plt.contourf(A, B, l)
plt.contour(A, B, l, levels=100)
plt.colorbar()
```

```python
l = loss(X[0], Y[0], A, B)
l += loss(X[1], Y[1], A, B)
l += loss(X[2], Y[2], A, B)
#plt.contourf(A, B, l)
plt.contour(A, B, l, levels=100)
plt.colorbar()

theta1 = np.array((3.5, 1))
plt.plot(theta1[0], theta1[1], "ro")
```

```python
step = 0.1

def gradient(a, b, X, Y):
    AT = np.vstack([X, np.ones_like(X)])
    Y_hat = a*X + b
    return AT@(Y_hat-Y)
```

```python
gradient(3.5, 1, X, Y)
```

```python
As = []
Bs = []

theta = theta1
As.append(theta[0])
Bs.append(theta[1])

for i in range(100):
    J = gradient(theta[0], theta[1], X, Y)
    theta -= J*step
    As.append(theta[0])
    Bs.append(theta[1])
```

```python
l = loss(X[0], Y[0], A, B)
l += loss(X[1], Y[1], A, B)
l += loss(X[2], Y[2], A, B)
#plt.contourf(A, B, l)
plt.contour(A, B, l, levels=100)
plt.colorbar()

theta1 = np.array((3.5, 1))
plt.plot(theta1[0], theta1[1], "ro")
plt.plot(As, Bs, ".--")
```
