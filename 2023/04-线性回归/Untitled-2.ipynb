```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 2
b = 5
```

```python
def f(X):
    return a * X + b
```

```python
X = np.linspace(0, 1, 101)
```

```python
Y = f(X)
```

```python
E = np.random.randn(101)
Y_observed = Y + E
```

```python
plt.plot(X, Y_observed, ".")
```

```python
A = np.vstack([X, np.ones_like(X)]).T
A
```

```python
ATA = A.T @ A
ATA
```

```python
ATy = A.T @ Y_observed
ATy
```

```python
theta = np.linalg.solve(ATA, ATy)
```

```python
theta
```

```python
a_star, b_star = theta
```

```python
plt.plot(X, Y_observed, ".")
plt.plot(X, a_star * X + b_star)
```

```python
X = np.array([2.0,3.0,4.0])
Y = np.array([10.0, 10.0, 14.0])
```

```python
def E(a, b, X, Y):
    VE = a*X + b - Y
    return np.abs(VE).sum()
```

```python
A = np.linspace(-3, 7, 401)
B = np.linspace(2, 8, 401)
```

```python
AA, BB = np.meshgrid(A, B)
```

```python
Result = np.empty_like(AA)

for i in range(AA.shape[0]):
    for j in range(AA.shape[1]):
        a = AA[i, j]
        b = BB[i, j]
        e = E(a, b, X, Y)
        Result[i, j] = e
```

```python
plt.contour(AA, BB, Result, 50);
```

```python
def linear_regression(X, Y, n):
    A = make_A_matrix(X, n)
    ATA = A.T @ A
    ATy = A.T @ Y
    theta = np.linalg.solve(ATA, ATy)
    return theta

def make_A_matrix(X, n):
    rows = [X**i for i in range(n, -1, -1)]
    A = np.vstack(rows).T
    return A
```

```python
def draw_data_and_model(X, Y, theta):
    x_min = X.min()
    x_max = X.max()

    X_m = np.linspace(x_min, x_max, 101)
    Y_m = model(X_m, theta)
    plt.plot(X, Y, ".")
    plt.plot(X_m, Y_m)
    
def model(X, theta):
    total = np.zeros_like(X)
    degrees = reversed(range(len(theta)))
    for n, w in zip(degrees, theta):
        total += w * X**n
    return total
```

```python
draw_data_and_model(X, Y, theta)
```
